```python
import requests

url = 'https://support.microsoft.com/pt-br/topic/n%C3%BAmeros-de-telefone-globais-do-servi%C3%A7o-de-atendimento-ao-cliente-c0389ade-5640-e588-8b0e-28de8afeb3f2'
res = requests.get(url)
html_page = res.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_page, 'html.parser')
```

```python
soup.find_all("div", {"class" : "ocpExpandoBody"})
```

```python
import re
ocpExpandoBody = []
for index, ocpExpandoBody in enumerate(soup.find_all("div", {"class" : "ocpExpandoBody"}), start = 1):
    pais = ocpExpandoBody.find("p", {"banded": "pais"})
    idioma = ocpExpandoBody.find("p", {"banded" : "provincia"}) 
```
